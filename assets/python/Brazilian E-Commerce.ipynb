# Data Cleaning

## Customers Dataset

```python
df_customers
```

#### Removing (customer_unique_id, customer_city, customer_state)

```python
df_customers = df_customers[['customer_id', 'customer_zip_code_prefix']]
df_customers
```

## Geolocation Dataset

```python
df_geolocation
```

#### Removing Duplicates Rows

```python
df_geolocation = df_geolocation.drop_duplicates()
df_geolocation
```

#### Removing Duplicates geolocation_zip_code_prefix

```python
df_geolocation = df_geolocation.drop_duplicates(subset=['geolocation_zip_code_prefix'], keep='first')
df_geolocation = df_geolocation.reset_index(drop=True)
df_geolocation
```

## Orders Dataset

```python
df_orders
```

```python
df_orders.columns
```

#### Removing (order_approved_at, order_delivered_carrier_date)

```python
df_orders = df_orders[['order_id', 'customer_id', 'order_status', 'order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']]
df_orders
```

#### Change order_purchase_timestamp, order_delivered_customer_date, order_estimated_delivery_date to Datestamp

```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'], errors='coerce')
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'], errors='coerce')
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'], errors='coerce')
```

```python
df_orders['order_delivered_customer_date'] = df_orders['order_delivered_customer_date'].fillna(df_orders['order_estimated_delivery_date'] - pd.Timedelta(days=1))
df_orders.info()
```

## Products Dataset

```python
df_products
```

#### Removing product_name_lenght, product_description_lenght, product_photos_qty, product_weight_g, product_length_cm, product_height_cm, product_width_cm

```python
df_products = df_products[['product_id', 'product_category_name']]
```

## Sellers Dataset

```python
df_sellers
```

#### Removing seller_city, seller_state

```python
df_sellers = df_sellers[['seller_id', 'seller_zip_code_prefix']]
```

## Order Items Dataset

```python
df_order_items
```

#### Change shipping_limit_date to datestamp

```python
df_order_items['shipping_limit_date'] = pd.to_datetime(df_order_items['shipping_limit_date'])
```

#### Making Feature (Revenue)

```python
df_order_items.columns
```

```python
df_order_items['revenue'] = df_order_items['order_item_id'] * df_order_items['price']
```

## Order Reviews Dataset

```python
df_order_reviews
```

```python
df_order_reviews.columns
```

#### Removing review_comment_title, review_comment_message

```python
df_order_reviews = df_order_reviews[['review_id', 'order_id', 'review_score', 'review_creation_date', 'review_answer_timestamp']]
df_order_reviews
```

#### Change review_creation_date, review_answer_timestamp to datestamp

```python
df_order_reviews['review_creation_date'] = pd.to_datetime(df_order_reviews['review_creation_date'])
df_order_reviews['review_answer_timestamp'] = pd.to_datetime(df_order_reviews['review_answer_timestamp'])
df_order_reviews.info()
```

```python
df_customers.to_csv('customers.csv', index=False)
df_geolocation.to_csv('geolocation.csv', index=False)
df_order_items.to_csv('order_items.csv', index=False)
df_order_payments.to_csv('order_payments.csv', index=False)
df_order_reviews.to_csv('order_reviews.csv', index=False)
df_orders.to_csv('df_orders.csv', index=False)
df_product_category.to_csv('product_category.csv', index=False)
df_products.to_csv('products.csv', index=False)
df_sellers.to_csv('sellers.csv', index=False)
```
